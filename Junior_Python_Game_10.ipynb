```python
### 작동 소리 넣기1
from tkinter import *
import numpy as np
from PIL import Image, ImageTk
import time
from pygame import mixer
mixer.init()

angle_sound=mixer.Sound('angle_control.wav')
power_sound=mixer.Sound('power_up.wav')

win = Tk()
win_w, win_h = 1200, 700
win.geometry(f"{win_w}x{win_h}")

cvs = Canvas(win,width=win_w,height=win_h,bg="white",highlightthickness=0)
cvs.pack()

angle_up = False
angle_down = False

power_up = False
power_down = False

angle_sound_control=False
power_sound_control=False

def press(event):
    global angle_up,angle_down,power_up,power_down,angle_sound_control,power_sound_control
    if angle_sound_control==False:
        angle_sound_control=True
        if event.keysym=='Up':
            angle_up = True
            angle_sound.play()
        elif event.keysym=='Down':
            angle_down = True
            angle_sound.play()
    else:
        if event.keysym=='Up':
            angle_up = True
        elif event.keysym=='Down':
            angle_down = True      
            
    if power_sound_control==False:
        power_sound_control=True
        if event.keysym=='Right':
            power_up=True
            power_sound.play()
        elif event.keysym=='Left':
            power_down=True
            power_sound.play()
    else:
        if event.keysym=='Right':
            power_up=True
        elif event.keysym=='Left':
            power_down=True     
    
def release(event):
    global angle_up, angle_down, power_up, power_down,angle_sound_control,power_sound_control
    angle_sound.stop()
    angle_sound_control=False
    power_sound.stop()
    power_sound_control=False 
    
    if event.keysym=='Up':
        angle_up=False
    elif event.keysym=='Down':
        angle_down=False
    elif event.keysym=='Right':
        power_up=False
    elif event.keysym=='Left':
        power_down=False       
    
win.bind("<KeyPress>", press)
win.bind("<KeyRelease>", release)


# 높이 값
bottom_h = round(win_h/6) # 150
ground_h = round(bottom_h/2)
h = bottom_h + ground_h

# 바닥
bottom = cvs.create_rectangle(0,win_h-bottom_h, win_w,win_h, fill="#bfbfbf", outline="#bfbfbf")

# 하늘
sky = cvs.create_rectangle(0,0, win_w,win_h-h, fill="#b3e6ff", outline="#b3e6ff")

# 산
p_mount = (400,win_h-h),(500,win_h-h-70),(550,win_h-h-30),(650,win_h-h-150),(730,win_h-h-40),(780,win_h-h-80),(880,win_h-h),(400,win_h-h)
mount = cvs.create_polygon(p_mount,fill="#40bf40",outline="#40bf40")

# 대포
angle_min = 10*np.pi/180
angle_max = 85*np.pi/180
angle = 45*np.pi/180
barrel = cvs.create_line(50,win_h-h, 50+round(130*np.cos(angle)),win_h-h-round(130*np.sin(angle)),
                        fill="#404040",width=30)
arc1 = cvs.create_arc(0, win_h-h-50, 100, win_h-h+50, fill="#808080", extent=180)
arc2 = cvs.create_arc(30, win_h-h-20, 70, win_h-h+20, fill="#404040", extent=180)

# 땅
ground = cvs.create_rectangle(0,win_h-h, win_w,win_h-bottom_h, fill="#b33c00", outline="#b33c00")

# 각도 게이지: 160x160 박스안의 반원(반지름 80)
arc_meter_1 = cvs.create_arc(120, win_h-bottom_h+30, 280, win_h-bottom_h+190, fill="beige", extent=180)
needle = cvs.create_line(200, win_h-bottom_h+110, 200+round(76*np.cos(angle)), win_h-bottom_h+110-round(76*np.sin(angle)),
                        fill="blue",width=4)
arc_meter_2 = cvs.create_arc(190, win_h-bottom_h+100, 210, win_h-bottom_h+120, fill="red", extent=180)
arc_meter_label = cvs.create_text(200,win_h-bottom_h+18,text="ANGLE",font=("Arial",15))

# 파워 게이지: 160x80 박스
power_min=10
power_max=200
power=50
power_meter_1 = cvs.create_rectangle(500,win_h-80, 710, win_h-5, fill="beige")#, outline="#404040", width=10)
power_meter_2 = cvs.create_rectangle(505,win_h-75, 505+power, win_h-10, fill="red",outline="red")
power_meter_label = cvs.create_text(600,win_h-bottom_h+18,text="POWER",font=("Arial",15))

# 시간: 1050
r_time = 25
time_board = cvs.create_rectangle(920,win_h-80, 1080, win_h-5, fill="beige")
time_value = cvs.create_text(1000,win_h-42,text=str(r_time),fill="red",font=("Arial",40))
time_label = cvs.create_text(1000,win_h-bottom_h+18,text="TIME",font=("Arial",15))

# 점수판: 60x60
score = 120
score_value = cvs.create_text(50,50,text=str(score),fill="black",font=("Arial",30))
score_label = cvs.create_text(50,15,text="SCORE",font=("Arial",12))

# 우주선
img2 = Image.open('spaceship_s3.png')
img2 = ImageTk.PhotoImage(img2)

ufo_x = 1100
ufo_y = 300
ufo = cvs.create_image(ufo_x, ufo_y,image=img2)  


while True:
    cvs.delete(barrel)
    cvs.delete(arc1)
    cvs.delete(arc2)
    cvs.delete(ground)
    cvs.delete(needle)
    cvs.delete(arc_meter_2)
    cvs.delete(power_meter_1)
    cvs.delete(power_meter_2)

    if angle_up==True and angle<angle_max:
        angle += 0.1
    elif angle_down==True and angle>angle_min:
        angle -= 0.1
        
    if power_up==True and power< power_max:
        power += 5
    elif power_down==True and power> power_min:
        power -= 5        
        
    barrel = cvs.create_line(50,win_h-h, 50+round(130*np.cos(angle)),win_h-h-round(130*np.sin(angle)),
                            fill="#404040",width=30)
    arc1 = cvs.create_arc(0, win_h-h-50, 100, win_h-h+50, fill="#808080", extent=180)
    arc2 = cvs.create_arc(30, win_h-h-20, 70, win_h-h+20, fill="#404040", extent=180)    
    ground = cvs.create_rectangle(0,win_h-h, win_w,win_h-bottom_h, fill="#b33c00", outline="#b33c00")

    needle = cvs.create_line(200, win_h-bottom_h+110, 200+round(76*np.cos(angle)), win_h-bottom_h+110-round(76*np.sin(angle)),
                        fill="blue",width=4)
    arc_meter_2 = cvs.create_arc(190, win_h-bottom_h+100, 210, win_h-bottom_h+120, fill="red", extent=180)    

    power_meter_1 = cvs.create_rectangle(500,win_h-80, 710, win_h-5, fill="beige")#, outline="#404040", width=10)
    power_meter_2 = cvs.create_rectangle(505,win_h-75, 505+power, win_h-10, fill="red",outline="red")    
    
    time.sleep(0.1)
    win.update()
    
win.mainloop()
```

```python
### 표적 명중 판단: 명중시 폭발 소리 넣기, 폭발 이미지 넣기
from tkinter import *
import numpy as np
from PIL import Image, ImageTk
import time
from pygame import mixer
mixer.init()

angle_sound = mixer.Sound('angle_control.wav')
power_sound = mixer.Sound('power_up.wav')
fire_sound = mixer.Sound('cannon_fire1.wav')
hit_sound = mixer.Sound('expl3.wav')

win = Tk()
win_w, win_h = 1200, 700
win.geometry(f"{win_w}x{win_h}")

cvs = Canvas(win,width=win_w,height=win_h,bg="white",highlightthickness=0)
cvs.pack()

angle_up = False
angle_down = False
angle_sound_control = False

power_up = False
power_down = False
power_sound_control = False

bomb_go = False
hit = False

def press(event):
    global angle_up, angle_down, angle_sound_control, power_up, power_down, power_sound_control,\
    bomb_go, bomb_x, bomb_y, bomb_v_x, bomb_v_y
    
    if event.keysym=='Up':
        if angle_sound_control==False:
            angle_up=True
            angle_sound.play()      
            angle_sound_control=True
        else:
            angle_up=True
            
    elif event.keysym=='Down':
        if angle_sound_control==False:
            angle_down=True
            angle_sound.play()         
            angle_sound_control=True
        else:
            angle_down=True    
            
    if event.keysym=='Right':
        if power_sound_control==False:
            power_up=True
            power_sound.play()      
            power_sound_control=True
        else:
            power_up=True
            
    elif event.keysym=='Left':
        if power_sound_control==False:
            power_down=True
            power_sound.play()      
            power_sound_control=True
        else:
            power_down=True    
            
    elif event.keysym=='space' and bomb_go==False:
        bomb_go=True
        fire_sound.play()     
        
        bomb_v_x = 40*power/185*np.cos(angle)
        bomb_v_y = 40*power/185*np.sin(angle)
        bomb_x = 50+round(130*np.cos(angle))-bomb_v_x
        bomb_y = win_h-h-round(130*np.sin(angle))+(bomb_v_y-10)            
            
def release(event):
    global angle_up, angle_down, power_up, power_down,angle_sound_control,power_sound_control
    angle_sound.stop()
    angle_sound_control=False
    power_sound.stop()
    power_sound_control=False 
    
    if event.keysym=='Up':
        angle_up=False
    elif event.keysym=='Down':
        angle_down=False
    elif event.keysym=='Right':
        power_up=False
    elif event.keysym=='Left':
        power_down=False               
         
win.bind("<KeyPress>", press)
win.bind("<KeyRelease>", release)

# 높이 값
bottom_h = round(win_h/6) # 150
ground_h = round(bottom_h/2)
h = bottom_h + ground_h

# 바닥
bottom = cvs.create_rectangle(0,win_h-bottom_h, win_w,win_h, fill="#bfbfbf", outline="#bfbfbf")

# 하늘
sky = cvs.create_rectangle(0,0, win_w,win_h-h, fill="#b3e6ff", outline="#b3e6ff")

# 산
p_mount = (400,win_h-h),(500,win_h-h-70),(550,win_h-h-30),(650,win_h-h-150),(730,win_h-h-40),(780,win_h-h-80),(880,win_h-h),(400,win_h-h)
mount = cvs.create_polygon(p_mount,fill="#40bf40",outline="#40bf40")

# 대포
angle_min = 10*np.pi/180
angle_max = 85*np.pi/180
angle = 45*np.pi/180
barrel = cvs.create_line(50,win_h-h, 50+round(130*np.cos(angle)),win_h-h-round(130*np.sin(angle)),
                        fill="#404040",width=30)
arc1 = cvs.create_arc(0, win_h-h-50, 100, win_h-h+50, fill="#808080", extent=180)
arc2 = cvs.create_arc(30, win_h-h-20, 70, win_h-h+20, fill="#404040", extent=180)

# 땅
ground = cvs.create_rectangle(0,win_h-h, win_w,win_h-bottom_h, fill="#b33c00", outline="#b33c00")

# 각도 게이지: 160x160 박스안의 반원(반지름 80)
arc_meter_1 = cvs.create_arc(120, win_h-bottom_h+30, 280, win_h-bottom_h+190, fill="beige", extent=180)
needle = cvs.create_line(200, win_h-bottom_h+110, 200+round(76*np.cos(angle)), win_h-bottom_h+110-round(76*np.sin(angle)),
                        fill="blue",width=4)
arc_meter_2 = cvs.create_arc(190, win_h-bottom_h+100, 210, win_h-bottom_h+120, fill="red", extent=180)
arc_meter_label = cvs.create_text(200,win_h-bottom_h+18,text="ANGLE",font=("Arial",15))

# 파워 게이지: 160x80 박스
power_min=10
power_max=200
power=50
power_meter_1 = cvs.create_rectangle(500,win_h-80, 710, win_h-5, fill="beige")#, outline="#404040", width=10)
power_meter_2 = cvs.create_rectangle(505,win_h-75, 505+power, win_h-10, fill="red",outline="red")
power_meter_label = cvs.create_text(600,win_h-bottom_h+18,text="POWER",font=("Arial",15))

# 시간: 1050
r_time = 25
time_board = cvs.create_rectangle(920,win_h-80, 1080, win_h-5, fill="beige")
time_value = cvs.create_text(1000,win_h-42,text=str(r_time),fill="red",font=("Arial",40))
time_label = cvs.create_text(1000,win_h-bottom_h+18,text="TIME",font=("Arial",15))

# 점수판: 60x60
score = 120
score_value = cvs.create_text(50,50,text=str(score),fill="black",font=("Arial",30))
score_label = cvs.create_text(50,15,text="SCORE",font=("Arial",12))

# 우주선
img1 = Image.open('spaceship_s3.png')
img1 = ImageTk.PhotoImage(img1)

ufo_x = 1100
ufo_y = 300
ufo = cvs.create_image(ufo_x, ufo_y,image=img1)  

# 포탄
img2 = Image.open('red_ball_s2.png')
img2 = ImageTk.PhotoImage(img2)

# 폭발
img3 = Image.open('explosion.png')
img3 = ImageTk.PhotoImage(img3)

win.update()

while True:
    cvs.delete(barrel)
    cvs.delete(arc1)
    cvs.delete(arc2)
    cvs.delete(ground)
    cvs.delete(needle)
    cvs.delete(arc_meter_2)
    cvs.delete(power_meter_1)
    cvs.delete(power_meter_2)
    
    try:
        cvs.delete(bomb)
        cvs.delete(expl)        
    except:
        pass    
    
    if angle_up==True and angle<=angle_max:
        angle += 0.01
    elif angle_down==True and angle>=angle_min:
        angle -= 0.01
        
    if power_up==True and power< power_max:
        power += 1
    elif power_down==True and power> power_min:
        power -= 1
        
    if bomb_go == True:
        bomb_x += bomb_v_x
        bomb_v_y += -1
        bomb_y -= bomb_v_y
        bomb = cvs.create_image(bomb_x, bomb_y,image=img2)   
        
        dist = ((ufo_x - bomb_x)**2 + (ufo_y - bomb_y)**2)**0.5
        if dist < 45:
            hit = True
            hit_sound.play()
            expl=cvs.create_image(bomb_x,bomb_y,image=img3)
        
        if bomb_y >= win_h-h or bomb_x >= win_w or hit:
            bomb_go=False
            hit=False
                   
    barrel = cvs.create_line(50,win_h-h, 50+round(130*np.cos(angle)),
                            win_h-h-round(130*np.sin(angle)),
                            fill="#404040",width=30)
    arc1 = cvs.create_arc(0, win_h-h-50, 100, win_h-h+50, fill="#808080", extent=180)
    arc2 = cvs.create_arc(30, win_h-h-20, 70, win_h-h+20, fill="#404040", extent=180)    
    ground = cvs.create_rectangle(0,win_h-h, win_w,win_h-bottom_h, fill="#b33c00",
                                  outline="#b33c00")

    needle = cvs.create_line(200, win_h-bottom_h+110, 200+round(76*np.cos(angle)),
                             win_h-bottom_h+110-round(76*np.sin(angle)),
                        fill="blue",width=4)
    arc_meter_2 = cvs.create_arc(190, win_h-bottom_h+100, 210, win_h-bottom_h+120,\
                                 fill="red", extent=180)    
    power_meter_1 = cvs.create_rectangle(500,win_h-80, 710, win_h-5, fill="beige")#, outline="#404040", width=10)
    power_meter_2 = cvs.create_rectangle(505,win_h-75, 505+power, win_h-10,\
                                         fill="red",outline="red")    

    time.sleep(0.01)
    win.update()
    
win.mainloop()
```
